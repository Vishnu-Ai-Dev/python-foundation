```python
#using the numpy building the mini ml predicting model using the linear regression 

import numpy as np
hours=np.array([1,2,3,4,5,6])
marks=np.array([30,40,50,60,70])
slope=(marks[-1]-marks[0])/(hours[-1]/hours[0])
intercept=marks[0]-slope*hours[0]
new_hours=6
predicted_marks=slope*new_hours+intercept
print("predicted marks for 6 hours is:",predicted_marks)
```
